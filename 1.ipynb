```python
#Question:03
n = int(input("Enter n: "))
numbers = []

print(str(n) + " numbers:")
for i in range(n):
    num = int(input())
    numbers.append(num)

for num in numbers:
    print(num)

total = 0
for num in numbers:
    total += num
print("Sum:", total)

minimum = numbers[0]
maximum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Maximum:", maximum)
print("Minimum:", minimum)

new=int(input("Enter new number: "))
numbers.append(new)
print("Updated list: ",numbers)
```

```python
#Question:04
amount=int(input("Enter purchase amount: "))
print("price: ",amount)
def discount(amount):
    if amount<1000:
        discount =5
    elif amount<5000:
        discount=10
    else:
        discount=15

    newamount=amount-(discount*amount /100)
    return newamount

newprice=discount(amount)
print("Discounted price: ",newprice)
```

```python
#Question:05
def counter(string):
    vowels = "aeiouAEIOU"
    vowels_count = 0
    consonants_count = 0

    for ch in string:
        if ch.isalpha():
            if ch in vowels:
                vowels_count += 1
            else:
                consonants_count += 1

    return {
        "Total Characters": len(string),
        "Total Vowels": vowels_count,
        "Total Consonants": consonants_count
    }

text = input("Enter a string: ")
print(text)
result = counter(text)
print(result)
```

```python
#Question:06
s = input("Enter string: ")
print("String: ",s)
longest = ""

for i in range(len(s)):
    temp = ""
    for j in range(i, len(s)):
        if s[j] not in temp:
            temp = temp + s[j]
        else:
            break
    if len(temp) > len(longest):
        longest = temp

print("Longest substring:", longest)
```

```python
#Question:07
numbers=input("Enter number: ").split()
print(numbers)
def find_repeats(numbers):
    repeated = []  

    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] == numbers[j] and numbers[i] not in repeated:
                repeated.append(numbers[i])
                break 

    return repeated

print("Repeated numbers:", find_repeats(numbers))
```
